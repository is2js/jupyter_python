# 1. 필요없는 칼럼 빼고, 분류를 인덱스로 주기

```python
muscle = raw_data.drop(['부위', '통증부위'], axis=1)
```

```python
muscle
```

```python
muscle = muscle.set_index(['분류'])
```

```python
muscle
```

# 2. str.get_dummies를 이용하여  해당근육을 나누기

```python
# 콤마로 나누기전에 빈칸없애기
muscle['해당근육'] = muscle['해당근육'].str.replace(' ', '')
```

```python
#콤마를 통해, 근육리스트들 나누기
muscle_token = muscle['해당근육'].str.get_dummies(sep=',')
muscle_token
```

```python
muscle_token.columns
```

```python
#분류같은 것들은 하나로 통합하기
mps = muscle_token.loc[muscle_token.index=='MPS', :].sum(axis=0)
acp = muscle_token.loc[muscle_token.index=='경혈', :].sum(axis=0)
```

```python
mps
```

```python
# 만약 중복된 것들은 2이상로 세지말고 1로 세기
mps.loc[ mps >= 2, ]=1
mps
```

```python
#다시 경혈까지 정리하기
mps = muscle_token.loc[muscle_token.index=='MPS', :].sum(axis=0)
acp = muscle_token.loc[muscle_token.index=='경혈', :].sum(axis=0)

mps.loc[ mps >= 2, ]=1
acp.loc[ acp >= 2, ]=1
```

```python
mps
```

```python
acp
```

```python
# 2개의 시리즈 합치기
token = pd.DataFrame({
    'MPS' : mps,
    '경혈' : acp,
})
```

```python
token.head()
```

```python
token['중복여부'] = token['MPS']+token['경혈']
```

```python
token.head()
```

```python
token.loc[token['중복여부'] == 1,'중복여부'] = False
token.loc[token['중복여부'] >1, '중복여부'] = True
```

```python
token.head()
```

```python
# 전체 갯수
total = len(token.index)
# 중복근육 갯수
duple = token['중복여부'].sum()
# 중복 안한 근육 갯수
noduple = total - duple


print('************************************************************')
print('경혈과 MPS의 근육 중복율은 {}%입니다.'.format( round((duple/total)*100,2) ))
print('************************************************************')

print('전체근육 리스트는 {}'.format( token.index.tolist() ))

du_list = token.loc[token['중복여부']==True,].index.tolist()
print('중복근육 리스트는 {}'.format(  du_list ))
```

```python
du_list.plot()
```
